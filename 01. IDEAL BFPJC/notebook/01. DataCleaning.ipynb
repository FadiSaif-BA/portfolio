```python
district_cols = ['district_abyan', 'district_lahj', 'district_aldale']

#fill first district column with values from the two other columns into the first one, then assign to new column 'district'
dataset['district'] = (
    dataset[district_cols[0]]
    .fillna(dataset[district_cols[1]])
    .fillna(dataset[district_cols[2]])
)

#drop old district columns
dataset = dataset.drop(columns=district_cols)
```

```python
df = dataset

#get the current list of columns
current_cols = df.columns.tolist()

#define the columns you're manipulating
column_to_move = 'district'
anchor_column = 'governorate'

#remove the column from its current position
try:
    current_cols.remove(column_to_move)
except ValueError:
    pass

#find the index of the anchor column and calculate the insertion point
insert_index = current_cols.index(anchor_column) + 1

#insert the column back into the desired position
current_cols.insert(insert_index, column_to_move)

#re-index the DataFrame using the new list order
df = df[current_cols]

dataset = df
```

```python
# --- Unpivot the data ---

# First, define your 'ID' columns. These are the columns
# that identify a respondent and should NOT be unpivoted.
# Get these names from the 'new_cols' list you just made.
id_vars = [
        'consent', 'gps_coordinates', 'latitude', 'longitude', 'altitude',
        'precision', 'enumerator', 'date', 'governorate', 'district','village',
        'respondent'
]

# Melt the DataFrame
# 'id_vars' are the columns to keep
# 'var_name' is the name of the new column that holds the question names
# 'value_name' is the name of the new column that holds the answers
melted_df = dataset.melt(
    id_vars=id_vars, 
    var_name='question', 
    value_name='answer'
)

# Show the first few rows of the new, long-format data
melted_df.shape
```

```python
melted_df
```

```python
import numpy as np

df = melted_df #assign melted dataframe to a temporary variable
name_column = 'respondent'

#generate consistent unique ids
codes, unique_names = pd.factorize(df[name_column])

# We add 1 to the codes so the IDs start at 1 instead of 0 (more user-friendly)
df['respondent_id'] = codes + 1 

# Create a small mapping dictionary to show the consistent relationship
mapping_dict = dict(zip(unique_names, codes + 1))
print(list(mapping_dict.items())[:5]) # Show the first 5 mappings

#reorder and drop names columns

#reorder: Move the new 'Respondent_ID' to the front
current_cols = df.columns.tolist()
current_cols.remove('respondent_id')
current_cols.insert(0, 'respondent_id')
df = df[current_cols]

#drop the sensitive name column
df = df.drop(columns=[name_column])
```
